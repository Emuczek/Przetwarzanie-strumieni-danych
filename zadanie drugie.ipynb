```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,31, 31)
y = np.random.uniform(-2,20,31)

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5), sharey="all")
axes.set_xticks(x, [str(int(val)) for val in x])
axes.bar(x, y, color ='yellow',width = 0.4)
axes.plot(x,y, "--", label='Temperatura', color="red")
axes.scatter(x,y, label='_Temperatura', color="red",)
axes.legend()
axes.grid(visible=True)
plt.xlabel("Dzie")
plt.ylabel("Temperatura")
plt.show()
```

```python
fig, axes = plt.subplot_mosaic(mosaic=[[1, 2], [1, 3]], figsize=(15, 5), layout="constrained")
color_arr = [np.random.rand(3,) for x in range(3)]
axes[2].bar(x, y, color=color_arr[0],width = 0.4)
axes[2].set_xlabel("Dzie")
axes[2].set_ylabel("Temperatura")
axes[2].set_xticks(x, [str(int(val)) for val in x])
axes[1].plot(x,y, label='Temperatura', color=color_arr[1])
axes[1].set_xlabel("Dzie")
axes[1].set_ylabel("Temperatura")
axes[1].set_xticks(x, [str(int(val)) for val in x])
axes[3].scatter(x,y, label='_Temperatura', color=color_arr[2])
axes[3].set_xlabel("Dzie")
axes[3].set_xticks(x, [str(int(val)) for val in x])
axes[3].set_ylabel("Temperatura")
plt.show()
```
